```python
from pickle import load, dump

with open('../Data/data.pkl','rb') as file:
    X_train, y_train = load( file )
    X_test, y_test = load( file )
```

```python
from pandas import concat

X_test_normal = X_test[['loan_amount', 'income']]
X_test_to_map = X_test[['ffiec_msa_md_median_family_income', 'aus_GUS', 'lien_status_Subordinate Lien', 'total_units_100-149', 'total_units_2',
       'total_units_25-49', 'total_units_3', 'total_units_4',
       'total_units_5-24', 'total_units_50-99', 'total_units_>149',
       'conforming_loan_limit_NC', 'preapproval_Preapproval Requested']]

test = concat( [ y_test, X_test_normal ], axis=1 )

test['aus_GUS'] = X_test_to_map['aus_GUS'].map( { 0:'No', 1: 'Si' } )
test['lien_status'] = X_test_to_map['lien_status_Subordinate Lien'].map( { 0:'First Lien', 1: 'Subordinate Lien' } )
test['conforming_limit'] = X_test_to_map['conforming_loan_limit_NC'].map( { 0:'C', 1: 'NC' } )
test['preapproval'] = X_test_to_map['preapproval_Preapproval Requested'].map( { 0:'Preapproval Not Requested', 1: 'Preapproval Requested' } )
test['action_taken'] = test['action_taken'].map( { 0:'Negado', 1: 'Aceptado' } )
test['med_fam_income'] = X_test['ffiec_msa_md_median_family_income']

def create_single_column(row):
    if row.sum() == 0: return 1
    else: return row.idxmax().replace('total_units_', '')

total_units = X_test_to_map[['total_units_100-149', 'total_units_2',
       'total_units_25-49', 'total_units_3', 'total_units_4',
       'total_units_5-24', 'total_units_50-99', 'total_units_>149']]

test['total_units'] = total_units.apply(create_single_column, axis=1).astype(str)

test
```

```python
X_train_normal = X_train[['loan_amount', 'income']]
X_train_to_map = X_train[['ffiec_msa_md_median_family_income', 'aus_GUS', 'lien_status_Subordinate Lien', 'total_units_100-149', 'total_units_2',
       'total_units_25-49', 'total_units_3', 'total_units_4',
       'total_units_5-24', 'total_units_50-99', 'total_units_>149',
       'conforming_loan_limit_NC', 'preapproval_Preapproval Requested']]

train = concat( [ y_train, X_train_normal ], axis=1 )

train['aus_GUS'] = X_train_to_map['aus_GUS'].map( { 0:'No', 1: 'Si' } )
train['lien_status'] = X_train_to_map['lien_status_Subordinate Lien'].map( { 0:'First Lien', 1: 'Subordinate Lien' } )
train['conforming_loan_limit'] = X_train_to_map['conforming_loan_limit_NC'].map( { 0:'C', 1: 'NC' } )
train['preapproval'] = X_train_to_map['preapproval_Preapproval Requested'].map( { 0:'Preapproval Not Requested', 1: 'Preapproval Requested' } )
train['action_taken'] = train['action_taken'].map( { 0:'Negado', 1: 'Aceptado' } )
train['med_fam_income'] = X_train['ffiec_msa_md_median_family_income']

def create_single_column(row):
    if row.sum() == 0: return 1
    else: return row.idxmax().replace('total_units_', '')

total_units = X_train_to_map[['total_units_100-149', 'total_units_2',
       'total_units_25-49', 'total_units_3', 'total_units_4',
       'total_units_5-24', 'total_units_50-99', 'total_units_>149']]

train['total_units'] = total_units.apply(create_single_column, axis=1).astype(str)

train
```

```python
with open('../Data/data_eda.pkl','wb') as file:
    dump( train, file )
    dump( test, file )
```
